```python
#En la tarea de Clasificación anterior, se aseguró de estar trabajando con todos los datos numéricos; 
#sin embargo, los conjuntos de datos típicos no contienen solo valores numéricos. La mayoría de los datos 
#contendrán una combinación de datos numéricos y nominales, por lo que debemos entender cómo incorporar
#tanto cuando se trata de desarrollar modelos de regresión como de hacer predicciones.
#Las variables categóricas se pueden usar directamente como predictoras o variables predichas en un modelo de 
#regresión múltiple siempre que se hayan convertido a valores binarios. Para preprocesar los datos de 
#ventas según sea necesario, primero debemos convertir todas las variables de objeto en características 
#binarias que contengan clases "0" y "1". Afortunadamente, pandas tiene un método para crear estas 
#Variables ficticias (Dummy Variables) de la siguiente manera:
#Primero verifiquemos los tipos de datos de cada una de las variables:
credit.dtypes
```

```python
#Podemos ver que tenemos tres variables que se enumeran como tipo de datos de objeto; necesitamos 
#convertirlos en 'variables ficticias' para que puedan usarse en este problema de regresión múltiple.
#Nota: todas las variables que estamos convirtiendo son características; Si la variable dependiente 
#también entra en esta categoría, tendríamos que usar un proceso diferente llamado label Encoding y pronto verá por qué.
#Generación de las variables dummyes:
#Una vez que hemos identificado las variables object, podemos proceder con la generación de las variables 
#dummyes, esto permitirá convertir los datos no numericos a numericos, esto se realiza de la siguiente forma:
credit = pd.get_dummies(credit)
#La función pd.get_dummies(credit) en pandas se utiliza para realizar la codificación one-hot (binaria) de 
#las variables categóricas en un DataFrame. La codificación one-hot es un proceso en el que las variables 
#categóricas se convierten en variables binarias (0 o 1) para que puedan ser utilizadas en algoritmos de 
#aprendizaje automático que requieren entradas numéricas en lugar de categorías.
#Cuando ejecutas pd.get_dummies(credit), se toma el DataFrame credit como entrada y se realiza lo siguiente:
#Identifica todas las columnas que contienen variables categóricas (por ejemplo, variables que contienen 
#texto o categorías en lugar de valores numéricos).
#Para cada columna categórica, crea nuevas columnas binarias (0 o 1) para cada categoría presente en
#la columna original. Cada nueva columna representa una categoría y se establece en 1 si la observación
#pertenece a esa categoría y 0 en caso contrario.
#Elimina la columna original que contenía la variable categórica, ya que ha sido reemplazada por las nuevascolumnas binarias.
#El resultado de pd.get_dummies(credit) será un nuevo DataFrame donde todas las variables categóricas han 
#sido convertidas en variables binarias. Esto es útil para utilizar estos datos en algoritmos de aprendizaje 
#automático que requieren variables numéricas, como regresión lineal, árboles de decisión, regresión logística, 
#entre otros. La codificación one-hot ayuda a evitar problemas de interpretación incorrecta de las variables 
#categóricas como numéricas y permite que el modelo utilice adecuadamente la información categórica en el análisis.
#Observe cómo esto cambió la variable Educación (todas las columnas de 'objeto' se cambiaron de esta manera):
#Finalmente verifique nuevamente los tipos de datos de cada variable, esto para asegurar que todos son datos numericos.
```

```python
#SEMANA 10 (PCA)
credit = pd.read_excel ("Credit.xlsx")
credit.drop(['Unnamed: 0'], axis=1,inplace=True)
credit.head(3)
```

```python
credit.columns
```

```python
credit =credit.rename(columns= {'default_payment_next_month(Y)':'Y'})
#la columna en el DataFrame credit que anteriormente se llamaba 'default_payment_next_month(Y)' 
#ahora se llamará simplemente 'Y'.
```

```python
# Se define una lista llamada features que contiene los nombres de las columnas que se utilizarán 
#como características en el análisis.
features = ['LIMIT_BAL','AGE','PAY_0','PAY_2', 'PAY_3','PAY_4', 'PAY_5', 'PAY_6', 'BILL_AMT1', 'BILL_AMT2',
              'BILL_AMT3', 'BILL_AMT4','BILL_AMT5','BILL_AMT6', 'PAY_AMT1','PAY_AMT2','PAY_AMT3','PAY_AMT4',
              'PAY_AMT5','PAY_AMT6']
# Se seleccionan las columnas especificadas en features del DataFrame credit. 
# "x" contien las características, y "y" contiene la variable objetivo.
x = credit.loc[:, features].values
y = credit.loc[:,['Y']].values
# Se utiliza StandardScaler de scikit-learn para estandarizar (normalizar) las características en x. 
# La estandarización es un proceso que transforma las características para que tengan una media de 0
# y una desviación estándar de 1. Esto es útil para muchos algoritmos de aprendizaje automático que asumen
# que las características están en la misma escala.
x = StandardScaler().fit_transform(x)
print(x)
```

```python
#Se crea una instancia de la clase PCA de scikit-learn con la especificación de que se desean 
#retener solo las dos componentes principales (n_components=2).
pca = PCA(n_components=2)
#Se aplica el análisis de componentes principales a la matriz x utilizando el método fit_transform. 
#Este paso calcula las dos componentes principales para cada muestra en x.
principalComponets = pca.fit_transform(x)
#Los resultados del análisis de componentes principales se almacenan en un nuevo DataFrame llamado principalDf, 
# donde cada fila representa una muestra y las columnas 'principal_component_1' y 'principal_component_2' 
#representan las dos componentes principales.
principalDf = pd.DataFrame(data = principalComponets,columns = ['principal_component_1', 'principal_component_2'])
principalDf.head(10)
```

```python
finalDf = pd.concat([principalDf,credit[['Y']]], axis = 1)
finalDf.head(10)
#En resumen, este código crea un nuevo DataFrame llamado finalDf al concatenar horizontalmente
#las componentes principales (principalDf) obtenidas mediante PCA y la columna 'Y' de credit.
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 

ax.set_xlabel('Componente Principal 1', fontsize = 10)
ax.set_ylabel('Componente Principal 2', fontsize = 10)
ax.set_title('Reducción PCA (2 Componentes Principales)', fontsize = 14)

targets = ['default', 'not default']
colors = ['#686466','#96d7eb']

for target, color in zip(targets,colors):
    indicesToKeep = finalDf['Y'] == target
    ax.scatter(finalDf.loc[indicesToKeep, 'principal_component_1']
               , finalDf.loc[indicesToKeep, 'principal_component_2']
               , c = color
               , s = 50
               ,alpha=0.3
               ,edgecolors='none')
ax.legend(targets)
ax.grid()
```

```python
pca.explained_variance_ratio_
#Este código devuelve la proporción de la varianza explicada por cada componente principal 
#después de realizar un análisis de componentes principales (PCA) en los datos.
```

```python
viz = finalDf[['principal_component_1','principal_component_2','Y']]
viz.hist()
#En resumen, después de ejecutar estas líneas de código, se generará un conjunto de histogramas para las columnas
#'principal_component_1', 'principal_component_2' y 'Y' en el DataFrame finalDf
plt.show()
```
